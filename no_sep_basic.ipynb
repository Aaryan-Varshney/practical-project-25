```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import rc
# Enable LaTeX for text rendering
rc('text', usetex=True)
rc('text.latex', preamble="\\usepackage{siunitx}")
```

```python
initial_heights = np.array([
    0.804, 0.804, 0.8, 0.807, 0.802,
    0.561, 0.557, 0.563, 0.561, 0.55,
    0.741, 0.739, 0.745, 0.742,
    0.763, 0.76, 0.769, 0.763, 0.761,
    0.428, 0.431,
])

final_heights = np.array([
    1.848, 1.815, 1.867, 1.865, 1.7907,
    1.312, 1.23, 1.226, 1.283, 1.31,
    1.695, 1.65, 1.675, 1.65,
    1.811, 1.699, 1.775, 1.721, 1.719,
    0.957, 0.964,
])
```

```python
g = 9.81

approach_velocities = np.sqrt(2 * g * initial_heights)

separation_velocities = np.sqrt(2 * g * final_heights)

e_net = 1.52
```

```python
x = initial_heights
y = final_heights

xlims = (0.4, 0.9)
ylims = (0.8, 2.0)

cross_colour = '#317d44'
line_colour = '#7bc133'

xerror = 0.005
yerror = 0.005

# Predicted values using the equation h_1 = h_0 * e_net^2
predicted_x = np.arange(xlims[0], xlims[1] + 0.01, 0.01)
predicted_y = predicted_x * (e_net ** 2)
plt.plot(predicted_x, predicted_y, color='red', linewidth=0.5, label = 'Predicted values')

plt.errorbar(x, y, xerr=xerror, yerr=yerror, fmt='+', color=cross_colour)

plt.grid()

plt.xlim(xlims)
plt.ylim(ylims)

# Calculate the slope of the trendline passing through the origin
slope = np.sum(x * y) / np.sum(x * x)

# Generate trendline values
trendline_x = np.linspace(xlims[0], xlims[1], 100)
trendline_y = slope * trendline_x

plt.plot(trendline_x, trendline_y, color=line_colour, label='Line of best fit')

x_tick_positions = np.arange(xlims[0], xlims[1] + 0.01, 0.1)  # From 0.65 to 1.1 with step 0.05
plt.xticks(x_tick_positions)

y_tick_positions = np.arange(ylims[0], ylims[1] + 0.01, 0.2)  # From 0.5 to 0.85 with step 0.05
plt.yticks(y_tick_positions)

# Add gridlines
plt.grid(which='major', color='gray', linestyle='-', linewidth=0.8)  # Major gridlines
plt.minorticks_on()  # Enable minor ticks
plt.grid(which='minor', color='lightgray', linestyle='--', linewidth=0.5)  # Minor gridlines

# Predicted values with uncertainty
predicted_x = np.arange(xlims[0], xlims[1] + 0.01, 0.01)
predicted_y = predicted_x * (e_net - 0.08) **2
plt.plot(predicted_x, predicted_y, color='red', linewidth=0.5, label = 'Lower bound of prediction', linestyle = '--')

# Predicted values with uncertainty
predicted_x = np.arange(xlims[0], xlims[1] + 0.01, 0.01)
predicted_y = predicted_x * (e_net + 0.08)**2
plt.plot(predicted_x, predicted_y, color='red', linewidth=0.5, label = 'Upper bound of prediction', linestyle = '--')

plt.xlabel(r"$h_{\text{T}} / \si{\metre}$")
plt.ylabel(r"$h_{\text{T}}' / \si{\metre}$")

plt.legend()

plt.tight_layout()

plt.savefig(r"C:\Coding\practical-project-25\graphs\no-separation/ht'-vs-ht.png", dpi = 600)

plt.show()

print(slope)
```
