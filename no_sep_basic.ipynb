```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
initial_heights = np.array([
    0.804, 0.804, 0.8, 0.807, 0.802,
    0.561, 0.557, 0.563, 0.561, 0.55,
    0.741, 0.739, 0.745, 0.742,
    0.763, 0.76, 0.769, 0.763, 0.761,
    0.428, 0.431,
])

final_heights = np.array([
    1.848, 1.815, 1.867, 1.865, 1.7907,
    1.312, 1.23, 1.226, 1.283, 1.31,
    1.695, 1.65, 1.675, 1.65,
    1.811, 1.699, 1.775, 1.721, 1.719,
    0.957, 0.964,
])
```

```python
g = 9.81

approach_velocities = np.sqrt(2 * g * initial_heights)

separation_velocities = np.sqrt(2 * g * final_heights)

e_net = 1.52
```

```python
x = initial_heights
y = final_heights

xlims = (0.4, 0.9)
ylims = (0.8, 2.0)

# Predicted values using the equation h_1 = h_0 * e_net^2
predicted_x = np.arange(xlims[0], xlims[1] + 0.01, 0.01)
predicted_y = predicted_x * (e_net ** 2)
plt.plot(predicted_x, predicted_y, color='red')



plt.scatter(x, y)

plt.grid()

plt.xlim(xlims)
plt.ylim(ylims)

plt.show()
```

```python
x = approach_velocities
y = separation_velocities

print(min(x), max(x))
print(min(y), max(y))

xlims = (2.5, 4.5)
ylims = (4.0, 6.5)

# Predicted values using the equation v_1 = v_0 * e_net
predicted_x = np.arange(xlims[0], xlims[1] + 0.01, 0.01)
predicted_y = predicted_x * e_net
plt.plot(predicted_x, predicted_y, color='red')

plt.scatter(x, y, marker = "+")

plt.grid()

plt.xlim(xlims)
plt.ylim(ylims)

plt.show()
```
