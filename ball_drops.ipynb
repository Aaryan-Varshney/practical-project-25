```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
```

```python
initial = np.array([0.804, 0.804, 0.8, 0.807, 0.802, 0.561, 0.557, 0.563, 0.561, 0.55, 0.741, 0.739, 0.745, 0.742, 0.674, 0.672, 0.674, 0.673, 0.763, 0.76, 0.769, 0.763, 0.761])

final = np.array([1.848, 1.815, 1.867, 1.865, 1.7907, 1.312, 1.293, 1.226, 1.283, 1.31, 1.695, 1.712, 1.675, 1.71, 1.512, 1.49, 1.59, 1.64, 1.811, 1.699, 1.775, 1.721, 1.719])

x_range = (np.min(initial) - 0.05, np.max(initial) + 0.05)
y_range = (np.min(final) - 0.05, np.max(final) + 0.05)

x = initial
y = final

corr, _ = pearsonr(x, y)
```

```python
g = 9.81

v0 = np.sqrt(2 * g * initial)
v1 = np.sqrt(2 * g * final)

v0_range = (np.min(v0) - 0.5, np.max(v0) + 0.5)
v1_range = (np.min(v1) - 0.5, np.max(v1) + 0.5)
```

```python
# Plot the v0 graph
plt.figure(figsize=(6, 4))
plt.scatter(initial, v0, marker='x', color='#317d44')  # Same cross style
plt.xlabel(r'$\sqrt{h_0}$ / $m^{1/2}$')
plt.ylabel(r'$v_0$ / $m/s$')

# Calculate trendline for v0
slope_v0 = np.sum(initial * v0) / np.sum(initial * initial)
trendline_v0_x = np.linspace(np.min(initial), np.max(initial), 100)
trendline_v0_y = slope_v0 * trendline_v0_x
plt.plot(trendline_v0_x, trendline_v0_y, color='#7bc133')

plt.xlim(v0_range)
plt.ylim(v1_range)
plt.grid(which='major', color='gray', linestyle='-', linewidth=0.8)
plt.minorticks_on()
plt.grid(which='minor', color='lightgray', linestyle='--', linewidth=0.5)

plt.title('Velocity Before Bounce ($v_0$)')
plt.show()

# Plot the v1 graph
plt.figure(figsize=(6, 4))
plt.scatter(final, v1, marker='x', color='#317d44')  # Same cross style
plt.xlabel(r'$\sqrt{h_1}$ / $m^{1/2}$')
plt.ylabel(r'$v_1$ / $m/s$')

# Calculate trendline for v1
slope_v1 = np.sum(final * v1) / np.sum(final * final)
trendline_v1_x = np.linspace(np.min(final), np.max(final), 100)
trendline_v1_y = slope_v1 * trendline_v1_x
plt.plot(trendline_v1_x, trendline_v1_y, color='#7bc133')

plt.xlim(v0_range)
plt.ylim(v1_range)
plt.grid(which='major', color='gray', linestyle='-', linewidth=0.8)
plt.minorticks_on()
plt.grid(which='minor', color='lightgray', linestyle='--', linewidth=0.5)

plt.title('Velocity After Bounce ($v_1$)')
plt.show()
```

```python
# Define sqrt of initial and final heights for the x-axis
sqrt_initial = np.sqrt(initial)
sqrt_final = np.sqrt(final)

# Recalculate x and y ranges
x_v0_range = (np.min(sqrt_initial) - 0.05, np.max(sqrt_initial) + 0.05)
y_v0_range = (np.min(v0) - 0.5, np.max(v0) + 0.5)

# Plot the v0 graph
plt.figure(figsize=(6, 4))
plt.scatter(sqrt_initial, v0, marker='x', color='#317d44')  # Use sqrt(initial) for x-axis
plt.xlabel(r'$\sqrt{h_0}$ / $m^{1/2}$')
plt.ylabel(r'$v_0$ / $m/s$')

# Calculate trendline for v0
slope_v0 = np.sum(sqrt_initial * v0) / np.sum(sqrt_initial * sqrt_initial)
trendline_v0_x = np.linspace(np.min(sqrt_initial), np.max(sqrt_initial), 100)
trendline_v0_y = slope_v0 * trendline_v0_x
plt.plot(trendline_v0_x, trendline_v0_y, color='#7bc133')

# Set proper limits
plt.xlim(x_v0_range)
plt.ylim(y_v0_range)
plt.grid(which='major', color='gray', linestyle='-', linewidth=0.8)
plt.minorticks_on()
plt.grid(which='minor', color='lightgray', linestyle='--', linewidth=0.5)

plt.title('Velocity Before Bounce ($v_0$)')
plt.show()
```
