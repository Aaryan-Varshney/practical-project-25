```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
initial_heights = np.array([
    0.804, 0.807, 0.8, 0.8, 0.801, 0.804, 0.804, 0.807, 0.804
])

final_heights = np.array([
    1.9566144, 2.0401767, 1.97192, 2.12552, 2.2072356, 2.1624384, 2.1361476, 2.0918247, 1.93161
])

change_in_heights = final_heights - initial_heights

separation_distances = np.array([
    19, 14, 6, 29, 53, 36, 39, 27, 5
]) / 1000

# separation_distances = separation_distances * 0.01
```

```python
x = initial_heights
y = final_heights
z = separation_distances

xlims = (0, 1)
ylims = (0, 3)

plt.scatter(x, y, c=z, cmap='viridis')

# plt.xlim(xlims)
# plt.ylim(ylims)

plt.xlabel('Initial Height / m')
plt.ylabel('Final Height / m')
plt.colorbar(label='Separation Distances / m')
plt.grid(True)

plt.show()
```
