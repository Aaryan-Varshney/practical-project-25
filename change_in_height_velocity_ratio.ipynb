```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
```

```python
initial_heights = np.array([
    0.804, 0.804, 0.8, 0.807, 0.802,
    0.561, 0.557, 0.563, 0.561, 0.55,
    0.741, 0.739, 0.745, 0.742,
    0.674, 0.672, 0.674, 0.673,
    0.763, 0.76, 0.769, 0.763, 0.761,
    0.428, 0.431
])

final_heights = np.array([
    1.848, 1.815, 1.867, 1.865, 1.7907,
    1.312, 1.23, 1.226, 1.283, 1.31,
    1.695, 1.65, 1.675, 1.65,
    1.512, 1.49, 1.59, 1.64,
    1.811, 1.699, 1.775, 1.721, 1.719,
    0.957, 0.964
])

change_in_heights = final_heights - initial_heights
```

```python
g = 9.81

approach_velocities = np.sqrt(2 * g * initial_heights)

separation_velocities = np.sqrt(2 * g * final_heights)

velocity_ratios = separation_velocities / approach_velocities
```

```python
x = change_in_heights
y = velocity_ratios

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', edgecolor='black')
plt.title('Change in Height vs. Velocity Ratio')
plt.xlabel('Change in Height (m)')
plt.ylabel('Velocity Ratio (v_separation / v_approach)')
plt.grid(True)
plt.show()
```
