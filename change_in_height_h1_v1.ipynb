```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
initial_heights = np.array([
    0.804, 0.804, 0.8, 0.807, 0.802,
    0.741, 0.739, 0.745, 0.742,
    0.763, 0.76, 0.769, 0.763, 0.761,
])

final_heights = np.array([
    1.848, 1.815, 1.867, 1.865, 1.7907,
    1.695, 1.65, 1.675, 1.65,
    1.811, 1.699, 1.775, 1.721, 1.719,
])

change_in_heights = final_heights - initial_heights
```

```python
g = 9.81

approach_velocities = np.sqrt(2 * g * initial_heights)

separation_velocities = np.sqrt(2 * g * final_heights)

velocity_ratios = separation_velocities / approach_velocities
```

```python
x = change_in_heights
y1 = initial_heights

plt.scatter(x, y1)
# Calculate the slope of the trendline passing through the origin
slope = np.sum(x * y1) / np.sum(x * x)

# Generate trendline values
trendline_x = np.linspace(0, 1.5, 100)
trendline_y = slope * trendline_x

# Plot the trendline
plt.plot(trendline_x, trendline_y)


plt.title('Change in Height vs Initial Height')
plt.xlabel('Change in Height (m)')
plt.ylabel('Initial Height (m)')
plt.grid(True)
plt.show()

```

```python
x = change_in_heights
y2 = separation_velocities

plt.scatter(x, y2)
coefficients = np.polyfit(x, y2, 1)
slope, intercept = coefficients

# Generate trendline values
trendline_x = np.linspace(0, 1.1, 100)
trendline_y = slope * trendline_x + intercept

# Plot the trendline
plt.plot(trendline_x, trendline_y)

# Change axis limits
plt.xlim(0, 1.1)
plt.ylim(0, 7)

plt.title('Change in Height vs Separation Velocity')
plt.xlabel('Change in Height (m)')
plt.ylabel('Separation Velocity (m/s)')
plt.grid(True)
plt.show()
```
